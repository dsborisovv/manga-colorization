```python
import pandas as pd
import numpy as np
import os
import shutil
import json
```

Для начала сделаем датасет из 1000 сэмплов
Структура данных:
* /images - датасет с цветной мангой
* /conditioning_images - датасет с чб мангой
* /train.jsonl - метаданные. Пример - {"text": "aqua circle with light pink background", "image": "images/2.png", "conditioning_image": "conditioning_images/2.png"}

```python
# BW_PATH = 'D:/bw_data_512/'
# COLOR_PATH = 'D:/0/color_data_512/color_data_512/'
# IMAGES_PATH = 'D:/images/'
# CONDITIONING_IMAGES_PATH = 'D:/conditioning_images/'
# JSONL_PATH = 'D:/train.jsonl'
# IMAGES_NUM = 1_000

# для more_bw
BW_PATH = 'D:/more_bw/more_bw.tar/more_bw/25/'
COLOR_PATH = 'D:/0/color_data_512/color_data_512/'
IMAGES_PATH = 'D:/images_more_bw/'
CONDITIONING_IMAGES_PATH = 'D:/conditioning_images_more_bw/'
JSONL_PATH = 'D:/train_more_bw.jsonl'
IMAGES_NUM = 1_000
```

```python
def copy_image(image, path):
    shutil.copy(image, path)


def make_jsonl(data, jsonl_file_path):
    with open(jsonl_file_path, 'w') as jsonl_file:
        for item in data:
            jsonl_file.write(json.dumps(item) + '\n')
```

```python
data = []
cnt = 0

for bw_image in os.listdir(BW_PATH):
    if cnt == IMAGES_NUM:
        break
    elif cnt % 100 == 0:
        print(f'CNT: {cnt}')
        print()
    meta = {}
    for color_image in os.listdir(COLOR_PATH):
        if bw_image == color_image:
            copy_image(
                os.path.join(BW_PATH, bw_image),
                os.path.join(CONDITIONING_IMAGES_PATH, f'{cnt}.png')
            )
            copy_image(
                os.path.join(COLOR_PATH, color_image),
                os.path.join(IMAGES_PATH, f'{cnt}.png')
            )
            meta['text'] = 'color manga page'
            meta['image'] = f'images_more_bw/{cnt}.png'
            meta['conditioning_image'] = f'conditioning_images_more_bw/{cnt}.png'
            data.append(meta)
            cnt += 1
            break
```

```python
make_jsonl(data, JSONL_PATH)
```
